```python
import numpy as np
import pandas as pd

houseprice=pd.read_csv('AmesHousing.csv')
houseprice = houseprice.drop("PID", axis=1).drop("Order", axis=1)
houseprice.head()
```

```python
houseprice.info()
```

```python
print(f"rows: {houseprice.shape[0]}")
print(f"columns: {houseprice.shape[1]}")
```

```python
#check categories with missing values
missing_values = houseprice.isnull().sum()
print(missing_values[missing_values > 0])
```

```python
#impute missing values in numeric categories with the median
numeric_columns = houseprice.select_dtypes(include='number').columns
from sklearn.impute import SimpleImputer
imputerNum = SimpleImputer(strategy="median")
houseprice[numeric_columns] = imputerNum.fit_transform(houseprice[numeric_columns])
```

```python
#impute missing values in categorical categories as 'unknown'
categorical_columns = houseprice.select_dtypes(include='object').columns
imputerCat = SimpleImputer(strategy='constant', fill_value='unknown')
houseprice[categorical_columns] = imputerCat.fit_transform(houseprice[categorical_columns])
```

```python
#check missing values again to see there are none! yippieeee
missing_values = houseprice.isnull().sum()
print(missing_values[missing_values > 0])
```

```python
#print catergories and their specific values and count
categorical_columns = houseprice.select_dtypes(include='object').columns
for column in categorical_columns:
    print(houseprice[column].value_counts())
    print()
```
